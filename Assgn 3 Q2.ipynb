```python
equ = cv2.equalizeHist(gray)
#res = numpy.hstack((img,equ)) #stacking images side-by-side
cv2_imshow(equ)

max=gray[0][0]
min=gray[0][0]

for i in range(h):
  for j in range(w):
    if max<gray[i][j]:
      max=gray[i][j]
    if min>gray[i][j]:
      min=gray[i][j]

gray_1=gray.copy()

for i in range(h):
  for j in range(w):
    gray_1[i][j]=((gray[i][j]-min)*(255-0))/(max-min)

cv2_imshow(gray_1)
```
