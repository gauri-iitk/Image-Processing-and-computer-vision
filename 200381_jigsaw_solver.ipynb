```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
```

```python
#There are 4 sub-sections of the image which need to be changed, so plotting it to get the coordinates of those parts
image = cv.imread('/jigsaw.jpg') 
plt.imshow(image)
plt.show()
```

```python
#this part needs to be flipped about the y-axis
change1 = image[150:330,515:700]
image1 = cv.flip(change1,1)
```

```python
#this part needs to be flipped about the x-axis
change2 = image[370:420,370:795]
image2 = cv.flip(change2,0)
```

```python
#this part needs to be flipped about the x-axis
change3 = image[200:410,0:190]
image3 = cv.flip(change3,0) 
```

```python
#For this part, the color space has to be changed to GBR
change4 = image[0:200,0:190]
blue,green,red = cv.split(change4)
merged = cv.merge([green,blue,red])
image4 = merged
```

```python
#Finally putting all the parts to their correct locations
image[150:330,515:700] = image1
image[370:420,370:795] = image2
image[0:210,0:190] = image3
image[200:400,0:190] = image4
```

```python
cv2_imshow(image)
cv.waitKey(0)
```
